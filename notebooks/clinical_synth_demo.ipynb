This notebook demonstrates the initial phase of our clinical synthetic data generation framework using the liver disease dataset.

## üéØ Phase 1 Objectives
- Load and visualize the liver disease dataset
- Apply preprocessing pipeline
- Generate mock synthetic data from multiple models
- Create initial comparison visualizations
- Assess what the output tables and visualizations will look like

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sys
import os
from pathlib import Path

# Add src directory to path
sys.path.append('../src')

# Import our framework components
from preprocessing.clinical_preprocessor import ClinicalDataPreprocessor
from models.mock_models import MockCTGANModel, MockTVAEModel, MockCopulaGANModel, MockGANerAidModel

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seeds for reproducibility
np.random.seed(42)

print("‚úÖ All imports successful!")
print("üìÅ Framework components loaded")
```

## üìÇ 1. Data Loading and Initial Exploration

```python
# Load the liver disease dataset
DATA_FILE = "../doc/liver_train.csv"
TARGET_COLUMN = "result"  # Will be cleaned to 'result'
DATASET_NAME = "Liver Disease"

try:
    # Load the dataset
    df_original = pd.read_csv(DATA_FILE, encoding='ISO-8859-1')
    print(f"‚úÖ Dataset loaded successfully!")
    print(f"üìä Dataset: {DATASET_NAME}")
    print(f"üìè Original Shape: {df_original.shape}")
    
    # Display first few rows
    print("\nüìã First 5 rows:")
    display(df_original.head())
    
except FileNotFoundError:
    print(f"‚ùå Error: Could not find file {DATA_FILE}")
    raise
```

```python
# Basic dataset information
print("=" * 60)
print("üìã DATASET OVERVIEW")
print("=" * 60)

print(f"Dataset Shape: {df_original.shape}")
print(f"Memory Usage: {df_original.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print(f"Number of Features: {df_original.shape[1]}")
print(f"Number of Samples: {df_original.shape[0]}")

print("\nüìä Column Information:")
info_df = pd.DataFrame({
    'Column': df_original.columns,
    'Data_Type': df_original.dtypes,
    'Unique_Values': [df_original[col].nunique() for col in df_original.columns],
    'Missing_Values': [df_original[col].isnull().sum() for col in df_original.columns],
    'Missing_Percent': [df_original[col].isnull().sum()/len(df_original)*100 for col in df_original.columns]
})

display(info_df)

print("\nüéØ Target Variable Distribution:")
# Check different possible target column names
target_candidates = ['Result', 'result', 'RESULT']
target_col_found = None

for candidate in target_candidates:
    if candidate in df_original.columns:
        target_col_found = candidate
        break

if target_col_found:
    target_counts = df_original[target_col_found].value_counts()
    print(f"Target column found: {target_col_found}")
    print(target_counts)
    print(f"\nTarget Balance Ratio: {target_counts.min() / target_counts.max():.3f}")
    TARGET_COLUMN = target_col_found
else:
    print(f"‚ö†Ô∏è Warning: Target column not found. Available columns: {list(df_original.columns)}")
    TARGET_COLUMN = df_original.columns[-1]  # Use last column as target
    print(f"Using '{TARGET_COLUMN}' as target column")
```

### Initial Data Visualizations

```python
# Create comprehensive data visualizations
numeric_cols = df_original.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df_original.select_dtypes(include=['object']).columns.tolist()

# Remove target from feature lists
if TARGET_COLUMN in numeric_cols:
    numeric_cols.remove(TARGET_COLUMN)
if TARGET_COLUMN in categorical_cols:
    categorical_cols.remove(TARGET_COLUMN)

print(f"üìä Found {len(numeric_cols)} numeric and {len(categorical_cols)} categorical features")
print(f"üìä Numeric columns: {numeric_cols[:5]}{'...' if len(numeric_cols) > 5 else ''}")
print(f"üìä Categorical columns: {categorical_cols}")

# Plot distributions of numeric variables
if numeric_cols:
    n_cols = min(4, len(numeric_cols))
    n_rows = (len(numeric_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
    fig.suptitle('Distribution of Numeric Features', fontsize=16, y=1.02)
    
    if n_rows * n_cols == 1:
        axes = [axes]
    elif n_rows == 1:
        axes = axes
    else:
        axes = axes.flatten()
    
    for i, col in enumerate(numeric_cols):
        if i < len(axes):
            df_original[col].hist(bins=30, ax=axes[i], alpha=0.7, edgecolor='black')
            axes[i].set_title(f'{col}', fontsize=12)
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Frequency')
            axes[i].grid(True, alpha=0.3)
    
    # Remove empty subplots
    for j in range(len(numeric_cols), len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

# Plot categorical variables if any
if categorical_cols:
    n_cols = min(3, len(categorical_cols))
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
    fig.suptitle('Distribution of Categorical Features', fontsize=16, y=1.02)
    
    if n_rows * n_cols == 1:
        axes = [axes]
    elif n_rows == 1:
        axes = axes
    else:
        axes = axes.flatten()
    
    for i, col in enumerate(categorical_cols):
        if i < len(axes):
            value_counts = df_original[col].value_counts()
            value_counts.plot(kind='bar', ax=axes[i], color='skyblue', edgecolor='black')
            axes[i].set_title(f'{col}', fontsize=12)
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Count')
            axes[i].tick_params(axis='x', rotation=45)
            axes[i].grid(True, alpha=0.3)
    
    # Remove empty subplots
    for j in range(len(categorical_cols), len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
```

```python
# Correlation heatmap for numeric variables
if len(numeric_cols) > 1:
    plt.figure(figsize=(12, 10))
    
    # Include target if numeric
    cols_for_corr = numeric_cols.copy()
    if df_original[TARGET_COLUMN].dtype in ['int64', 'float64']:
        cols_for_corr.append(TARGET_COLUMN)
    
    correlation_matrix = df_original[cols_for_corr].corr()
    
    # Create heatmap
    sns.heatmap(correlation_matrix, 
                annot=True, 
                cmap='coolwarm', 
                center=0, 
                square=True, 
                linewidths=0.5,
                fmt='.2f')
    
    plt.title('Correlation Matrix of Numeric Variables', fontsize=16, pad=20)
    plt.tight_layout()
    plt.show()
    
    print("\nüîç Key Correlations with Target:")
    if TARGET_COLUMN in correlation_matrix.columns:
        target_corrs = correlation_matrix[TARGET_COLUMN].abs().sort_values(ascending=False)
        target_corrs = target_corrs[target_corrs.index != TARGET_COLUMN]
        print(target_corrs.head())
```

```python
# Initialize and apply preprocessing
preprocessor = ClinicalDataPreprocessor(random_state=42)

print("üîß Starting data preprocessing...")
print(f"Original data shape: {df_original.shape}")

# Apply preprocessing
df_processed = preprocessor.fit_transform(df_original, target_col=TARGET_COLUMN)

print(f"‚úÖ Preprocessing complete!")
print(f"Processed data shape: {df_processed.shape}")
print(f"Missing values after preprocessing: {df_processed.isnull().sum().sum()}")

# Show preprocessing summary
preprocessing_summary = preprocessor.generate_data_summary(df_processed)
print("\nüìã Preprocessing Summary:")
for key, value in preprocessing_summary.items():
    if key != 'column_types':
        print(f"  {key}: {value}")

print(f"\nüìã Column Types Identified:")
for col, col_type in preprocessing_summary['column_types'].items():
    print(f"  {col}: {col_type}")

# Identify discrete columns for synthetic data models
discrete_columns = preprocessor.get_discrete_columns(df_processed)
print(f"\nüè∑Ô∏è Discrete columns for models: {len(discrete_columns)}")
print(f"Discrete columns: {discrete_columns}")
```

```python
# Show before/after preprocessing comparison
print("\nüìä BEFORE vs AFTER Preprocessing Comparison:")
print("=" * 50)

comparison_df = pd.DataFrame({
    'Metric': ['Shape', 'Columns', 'Missing Values', 'Memory (MB)'],
    'Before': [
        str(df_original.shape),
        df_original.shape[1],
        df_original.isnull().sum().sum(),
        f"{df_original.memory_usage(deep=True).sum() / 1024**2:.2f}"
    ],
    'After': [
        str(df_processed.shape),
        df_processed.shape[1],
        df_processed.isnull().sum().sum(),
        f"{df_processed.memory_usage(deep=True).sum() / 1024**2:.2f}"
    ]
})

display(comparison_df)

# Display processed data sample
print("\nüìã Processed Data Sample:")
display(df_processed.head())
```

## ü§ñ 3. Mock Model Generation

Generate synthetic data using mock implementations of different models to demonstrate the framework.

```python
# Initialize mock models
models = {
    'CTGAN': MockCTGANModel(random_state=42),
    'TVAE': MockTVAEModel(random_state=42),
    'CopulaGAN': MockCopulaGANModel(random_state=42),
    'GANerAid': MockGANerAidModel(random_state=42)
}

print("ü§ñ Initializing mock models...")
print(f"Available models: {list(models.keys())}")

# Use a subset for faster demonstration
n_samples = min(1000, len(df_processed))
df_demo = df_processed.sample(n=n_samples, random_state=42)
print(f"\nüìä Using {n_samples} samples for demonstration")
print(f"Demo dataset shape: {df_demo.shape}")
```

```python
# Train models and generate synthetic data
synthetic_datasets = {}
model_performance = []

print("üöÄ Training models and generating synthetic data...")
print("=" * 60)

for model_name, model in models.items():
    print(f"\nüîß Training {model_name}...")
    
    try:
        # Create mock hyperparameters
        param_space = model.get_param_space()
        mock_params = {}
        for param, config in param_space.items():
            if config[0] == 'int':
                mock_params[param] = config[1] + (config[2] - config[1]) // 2
            elif config[0] == 'float':
                mock_params[param] = (config[1] + config[2]) / 2
            elif config[0] == 'categorical':
                mock_params[param] = config[1][0]
        
        # Create and fit model
        model.model = model.create_model(mock_params)
        model.fit(df_demo, discrete_columns)
        
        # Generate synthetic data
        synthetic_data = model.generate(len(df_demo))
        synthetic_datasets[model_name] = synthetic_data
        
        # Record performance metrics
        model_info = model.get_model_info()
        model_performance.append({
            'Model': model_name,
            'Training_Time': f"{model_info['training_time']:.3f}s",
            'Generation_Time': f"{model_info['generation_time']:.3f}s",
            'Samples_Generated': len(synthetic_data),
            'Features_Generated': len(synthetic_data.columns)
        })
        
        print(f"‚úÖ {model_name} completed - Generated {len(synthetic_data)} samples")
        
    except Exception as e:
        print(f"‚ùå {model_name} failed: {e}")
        continue

print(f"\nüéâ Successfully generated synthetic data from {len(synthetic_datasets)} models!")
```

```python
# Display model performance summary
performance_df = pd.DataFrame(model_performance)
print("üìä Model Performance Summary:")
print("=" * 60)
display(performance_df)

# Show sample of synthetic data from each model
print("\nüìã Synthetic Data Samples:")
print("=" * 60)

for model_name, synth_data in synthetic_datasets.items():
    print(f"\nü§ñ {model_name} Sample (first 3 rows):")
    display(synth_data.head(3))
```

## üìä 4. Initial Comparison Visualizations

```python
# Compare distributions between original and synthetic data
def plot_distribution_comparison(original_data, synthetic_datasets, columns_to_plot=None, max_cols=4):
    """Plot distribution comparisons between original and synthetic data."""
    
    if columns_to_plot is None:
        # Select numeric columns for comparison
        numeric_cols = original_data.select_dtypes(include=[np.number]).columns.tolist()
        columns_to_plot = numeric_cols[:max_cols]
    
    n_cols = len(columns_to_plot)
    n_models = len(synthetic_datasets)
    
    fig, axes = plt.subplots(n_cols, 1, figsize=(12, 4 * n_cols))
    if n_cols == 1:
        axes = [axes]
    
    colors = plt.cm.Set1(np.linspace(0, 1, n_models + 1))
    
    for i, col in enumerate(columns_to_plot):
        # Plot original data
        axes[i].hist(original_data[col], alpha=0.6, density=True, 
                    label='Original', color=colors[0], bins=30, edgecolor='black')
        
        # Plot synthetic data from each model
        for j, (model_name, synth_data) in enumerate(synthetic_datasets.items()):
            if col in synth_data.columns:
                axes[i].hist(synth_data[col], alpha=0.6, density=True, 
                           label=f'{model_name}', 
                           color=colors[j+1], bins=30, histtype='step', linewidth=2)
        
        axes[i].set_title(f'Distribution Comparison: {col}', fontsize=14)
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Density')
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)
    
    plt.suptitle('Original vs Synthetic Data Distributions', fontsize=16, y=1.02)
    plt.tight_layout()
    plt.show()

# Plot distribution comparisons
numeric_cols_for_plot = [col for col in df_demo.columns if df_demo[col].dtype in ['int64', 'float64']]
numeric_cols_for_plot = [col for col in numeric_cols_for_plot if col != TARGET_COLUMN.lower()][:4]

if numeric_cols_for_plot:
    plot_distribution_comparison(df_demo, synthetic_datasets, numeric_cols_for_plot)
else:
    print("‚ö†Ô∏è No suitable numeric columns found for distribution comparison")
```

```python
# Statistical comparison summary
def compare_statistics(original_data, synthetic_datasets, target_col):
    """Compare basic statistics between original and synthetic data."""
    
    comparison_results = []
    
    # Get numeric columns (excluding target)
    numeric_cols = [col for col in original_data.columns 
                    if original_data[col].dtype in ['int64', 'float64'] and col != target_col]
    
    for col in numeric_cols[:3]:  # Limit to first 3 for demo
        orig_stats = {
            'Column': col,
            'Dataset': 'Original',
            'Mean': original_data[col].mean(),
            'Std': original_data[col].std(),
            'Min': original_data[col].min(),
            'Max': original_data[col].max(),
            'Median': original_data[col].median()
        }
        comparison_results.append(orig_stats)
        
        # Add synthetic data statistics
        for model_name, synth_data in synthetic_datasets.items():
            if col in synth_data.columns:
                synth_stats = {
                    'Column': col,
                    'Dataset': f'{model_name}',
                    'Mean': synth_data[col].mean(),
                    'Std': synth_data[col].std(),
                    'Min': synth_data[col].min(),
                    'Max': synth_data[col].max(),
                    'Median': synth_data[col].median()
                }
                comparison_results.append(synth_stats)
    
    return pd.DataFrame(comparison_results)

# Generate statistical comparison
target_col_clean = TARGET_COLUMN.lower().replace(' ', '_')
stats_comparison = compare_statistics(df_demo, synthetic_datasets, target_col_clean)

print("üìä Statistical Comparison (Original vs Synthetic):")
print("=" * 80)

# Display formatted results
for col in stats_comparison['Column'].unique():
    col_data = stats_comparison[stats_comparison['Column'] == col]
    print(f"\nüìà {col.upper()}:")
    display(col_data.round(3))
```

```python
# Target variable distribution comparison
target_col_clean = TARGET_COLUMN.lower().replace(' ', '_')

if target_col_clean in df_demo.columns:
    fig, axes = plt.subplots(1, len(synthetic_datasets) + 1, figsize=(4 * (len(synthetic_datasets) + 1), 4))
    
    # Original target distribution
    target_counts_orig = df_demo[target_col_clean].value_counts()
    axes[0].pie(target_counts_orig.values, labels=target_counts_orig.index, autopct='%1.1f%%')
    axes[0].set_title('Original\nTarget Distribution')
    
    # Synthetic target distributions
    for i, (model_name, synth_data) in enumerate(synthetic_datasets.items()):
        if target_col_clean in synth_data.columns:
            target_counts_synth = synth_data[target_col_clean].value_counts()
            axes[i+1].pie(target_counts_synth.values, labels=target_counts_synth.index, autopct='%1.1f%%')
            axes[i+1].set_title(f'{model_name}\nTarget Distribution')
    
    plt.suptitle('Target Variable Distribution Comparison', fontsize=16, y=1.08)
    plt.tight_layout()
    plt.show()
    
    # Numerical comparison of target distributions
    print("\nüéØ Target Variable Distribution Comparison:")
    print("=" * 50)
    
    target_comparison = pd.DataFrame({
        'Dataset': ['Original'] + list(synthetic_datasets.keys()),
        'Class_0_Count': [target_counts_orig.get(0, 0) if 0 in target_counts_orig.index else target_counts_orig.iloc[0]] + 
                         [synth_data[target_col_clean].value_counts().get(0, 0) if 0 in synth_data[target_col_clean].value_counts().index 
                          else synth_data[target_col_clean].value_counts().iloc[0] for _, synth_data in synthetic_datasets.items()],
        'Class_1_Count': [target_counts_orig.get(1, 0) if 1 in target_counts_orig.index else target_counts_orig.iloc[1]] + 
                         [synth_data[target_col_clean].value_counts().get(1, 0) if 1 in synth_data[target_col_clean].value_counts().index 
                          else synth_data[target_col_clean].value_counts().iloc[1] for _, synth_data in synthetic_datasets.items()]
    })
    
    target_comparison['Total'] = target_comparison['Class_0_Count'] + target_comparison['Class_1_Count']
    target_comparison['Class_0_Pct'] = target_comparison['Class_0_Count'] / target_comparison['Total'] * 100
    target_comparison['Class_1_Pct'] = target_comparison['Class_1_Count'] / target_comparison['Total'] * 100
    
    display(target_comparison.round(2))
```

## üìà 5. Mock Evaluation Results

Create mock evaluation results to demonstrate what the final comparison tables and visualizations will look like.

```python
# Generate mock evaluation results
np.random.seed(42)

mock_results = []
model_names = list(synthetic_datasets.keys())

for i, model_name in enumerate(model_names):
    # Create realistic mock scores with some variance
    base_similarity = 0.75 + np.random.normal(0, 0.1)
    base_utility = 0.80 + np.random.normal(0, 0.08)
    
    mock_result = {
        'Model': model_name,
        'Overall_Similarity': np.clip(base_similarity, 0.5, 0.95),
        'Univariate_Similarity': np.clip(base_similarity + np.random.normal(0, 0.05), 0.5, 0.95),
        'Bivariate_Similarity': np.clip(base_similarity + np.random.normal(0, 0.08), 0.4, 0.9),
        'Correlation_Similarity': np.clip(base_similarity + np.random.normal(0, 0.1), 0.3, 0.95),
        'Average_Utility': np.clip(base_utility, 0.6, 0.95),
        'TRTR': np.clip(base_utility + np.random.normal(0, 0.05), 0.6, 0.95),
        'TSTS': np.clip(base_utility + np.random.normal(0, 0.06), 0.55, 0.9),
        'TRTS': np.clip(base_utility + np.random.normal(0, 0.08), 0.5, 0.9),
        'TSTR': np.clip(base_utility + np.random.normal(0, 0.07), 0.5, 0.9),
        'Training_Time_Sec': np.random.uniform(2, 15),
        'Generation_Time_Sec': np.random.uniform(0.5, 3),
        'N_Trials': 20
    }
    
    # Calculate combined score
    similarity_weight = 0.6
    mock_result['Combined_Score'] = (
        similarity_weight * mock_result['Overall_Similarity'] + 
        (1 - similarity_weight) * mock_result['Average_Utility']
    )
    
    mock_results.append(mock_result)

# Create results DataFrame
mock_results_df = pd.DataFrame(mock_results)
mock_results_df = mock_results_df.sort_values('Combined_Score', ascending=False).round(4)

print("üèÜ MOCK MODEL COMPARISON RESULTS")
print("=" * 80)
print("\nüìä Overall Rankings:")

# Display ranking summary
ranking_cols = ['Model', 'Combined_Score', 'Overall_Similarity', 'Average_Utility', 'Training_Time_Sec']
ranking_summary = mock_results_df[ranking_cols].copy()
ranking_summary.index = range(1, len(ranking_summary) + 1)
display(ranking_summary)

print("\nüìà Detailed Evaluation Metrics:")
detailed_cols = ['Model', 'Univariate_Similarity', 'Bivariate_Similarity', 'Correlation_Similarity', 
                'TRTR', 'TSTS', 'TRTS', 'TSTR']
detailed_summary = mock_results_df[detailed_cols].copy()
detailed_summary.index = range(1, len(detailed_summary) + 1)
display(detailed_summary)
```
